 # Twitch Channel Data Analysis

This dataset contains the data of the channels who are using Twitch Livestreaming platform to stream games.


Documentary Data Analysis:

Channel: The title of the Twitch Channel

Watch Time: This represents the total amount of minutes spent watching the channel's videos

Stream Time:- The total amount of time spent broadcasting the content by the channel.

Peak Viewers:- Number of people who watched the stream

Average Viewers: The average number of people who watch a certain station.

Followers:- The total number of people who follow a specific channel.

Gained Followers:- The number of subscribers obtained by a channel throughout time.

Views Gained:- The increase in the number of views of content on a specific channel throughout time.

Partnered: Indicates whether the channel partnered with another channel in a video.

Mature: Indicates if the channel's material is intended for adults or children.

Language: Specifies the primary language of the channel.

# EDA

## The below pie chart shows the percentage of channels intended for adults or childern.

```python
twitch_mature=twitch_dt.groupby(['Mature']).count()
```

```python
twitch_mature.Channel
```

```python
plt.figure(figsize=(9,7))
plt.title("Percentage of users above and below 18(Mature)")
plt.pie(twitch_mature["Channel"],labels = twitch_mature.index, autopct='%1.0f%%',startangle=60)
plt.show()
```

## The below Bar chart shows the number of Channels based on their specific language.

```python
language_ct = twitch_dt["Language"].value_counts()
language_ct
```

```python
plt.figure(figsize=(25,8))
plot = sns.countplot(x= 'Language',data= twitch_dt,order = twitch_dt['Language'].value_counts().index)
```

## This shows the count English vs Other Languages.

```python
twitch_dt['is_english_channel'] = np.where(twitch_dt['Language']=='English', True, False)
```

```python
plt.figure(figsize=(8, 8))
plot = sns.countplot(x="is_english_channel", data=twitch_dt,order=twitch_dt['is_english_channel'].value_counts().index)
```

## The below bar plot shows the top 20 Channels based on there stream time(in minutes)

```python
twitch_dt.set_index('Channel',inplace=True)
```

```python
twitch_dt.head()
```

```python
plt.figure(figsize = (21,8))

twitch_dt.sort_values('Stream time(minutes)',ascending=False)['Stream time(minutes)'].head(20).plot.bar(color = 'red')
plt.title('different stream times of Players (in minutes)')
plt.xlabel('Game_Streamers')
plt.ylabel('Count_of_stream_time')
plt.show()
```

The below line plot shows the graph for streamtime vs views gained

```python
plt.figure(figsize=(15,7))
sns.lineplot(twitch_dt['Stream time(minutes)'], twitch_dt['Views gained'],color='gray')
plt.ylabel('Views gained in hundred million')
plt.title(' Stream-time v/s views-gained', fontsize = 25)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(twitch_dt['Watch time(Minutes)'], twitch_dt['Views gained'],color='gray')
plt.xlabel('Watch Time in billions')
plt.ylabel('Views gained in hundred million')
plt.title('Watch time v/s views-gained', fontsize = 25)
plt.show()
```

The below Scatter plot shows the followers gained of a channel based on views.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(twitch_dt['Views gained'], twitch_dt['Followers gained'],color='green')
plt.title(' views-gained v/s followers-gained', fontsize = 15)
plt.ylabel('Followers gained in million')
plt.show()
```

The Bar plot shows the top 10 channels based on followers count.

```python
twitch_dt.sort_values(by = ['Followers'], ascending = False).head(10).plot.bar(y = 'Followers',color='black')
plt.title("Top ten Channels with max no of followers")
plt.ylabel('Followers in million')
plt.show()
```

The Below Heatmap shows the correlation between each column of the data set.

```python
fig = plt.figure(figsize=(10,8))
sns.heatmap(twitch_dt.corr(),annot=True)
plt.title(" Correlations between the columns")
plt.show()
```
